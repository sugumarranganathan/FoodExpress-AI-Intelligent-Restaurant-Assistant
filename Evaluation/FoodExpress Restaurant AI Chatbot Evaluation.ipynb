```python
import gradio as gr

with gr.Blocks(title="FoodExpress Restaurant AI Chatbot") as demo:

    gr.Markdown("""
# 🍽️ FoodExpress Restaurant AI Chatbot

### AI Powered Restaurant Assistant

Ask any question about the restaurant menu, timings, offers or services.

---

""")

    question = gr.Textbox(
        label="Ask Your Question",
        placeholder="Example : Burger List",
        lines=2
    )

    with gr.Row():

        ask_btn = gr.Button(
            "🚀 Ask AI",
            variant="primary"
        )

        clear_btn = gr.Button("🗑 Clear")

    gr.Examples(

        examples=[

            ["Burger List"],

            ["Pizza List"],

            ["Dessert List"],

            ["Opening Hours"],

            ["Today's Special"],

            ["Home Delivery"],

            ["Restaurant Address"],

            ["Cold Drinks"]

        ],

        inputs=question

    )

    with gr.Row():

        answer = gr.Textbox(

            label="🤖 AI Answer",

            lines=18,

            interactive=False

        )

        report = gr.Textbox(

            label="📊 Evaluation Dashboard",

            lines=25,

            interactive=False

        )

    ask_btn.click(

        fn=chatbot_live,

        inputs=question,

        outputs=[answer, report]

    )

    clear_btn.click(

        lambda: ("", "", ""),

        outputs=[question, answer, report]

    )

    gr.Markdown("""

---



""")

demo.launch(
    share=True,
    debug=True
)
```
